# Periodic noise in image processing

Periodic noise, also known as periodic pattern noise or structured noise, refers to a type of noise in images that exhibits a repeating pattern. This type of noise can be caused by various factors, such as electronic interference, errors in image compression, or errors in image processing algorithms

## Impact 

Periodic noise can have a significant impact on the quality of an image and can make it difficult to extract meaningful information from the image. For example, in medical imaging, periodic noise can make it difficult to accurately identify and diagnose certain conditions. In other applications, such as surveillance or remote sensing, periodic noise can make it difficult to detect objects or features of interest.

## remove or reduce

### First 

One common method is called frequency domain filtering. This approach involves transforming the image into the frequency domain using techniques such as Fourier transform or wavelet transform, and then applying a filter to remove or reduce the effects of the noise.

### Second

Another method is called spatial domain filtering. This approach involves applying a filter directly to the pixels in the image in order to remove or reduce the effects of the noise. Common filters used for this purpose include median filter, mean filter, and anisotropic diffusion filter.

### Third 

A third method is to use machine learning based techniques such as deep learning, to remove periodic noise from images. This can be done by training a neural network to recognize patterns in the noise and then removing those patterns from the image.

### Frequency domain filtering

Frequency domain filtering is called as such because it involves transforming the image from the spatial domain, where the image is represented in terms of its pixels, into the frequency domain, where the image is represented in terms of its frequency components. The frequency domain is a mathematical representation of the image that describes how the intensity of the pixels varies over different spatial scales.

The Fourier transform and the wavelet transform are the most common techniques used to transform an image from the spatial domain to the frequency domain. The Fourier transform converts an image into a representation of the image in terms of sine and cosine functions,
while the wavelet transform converts an image into a representation of the image in terms of wavelets, which are functions with localized frequency content.

Once the image is in the frequency domain, it is then possible to apply a filter to the frequency components of the image in order to remove or reduce the effects of the noise. This is where the term frequency domain filtering comes from.

For example, a low-pass filter will remove high-frequency noise, while a high-pass filter will remove low-frequency noise. This approach is effective because the noise in an image often has a distinct frequency content that can be separated from the useful information in the image. By isolating the noise in the frequency domain and removing it, it is possible to improve the quality of the image and extract meaningful information from it.

```python
import numpy as np
from scipy.fftpack import fft2, ifft2
from matplotlib import pyplot as plt

# Read an image
image = plt.imread("image.jpg")

# Convert the image to grayscale
gray_image = np.dot(image[...,:3], [0.2989, 0.5870, 0.1140])

plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')
plt.show()
# Apply the FFT to the image
fourier_transform = fft2(gray_image)

# Shift the zero-frequency component to the center of the spectrum
fourier_transform = np.fft.fftshift(fourier_transform)

# Compute the magnitude of the Fourier transform
magnitude_spectrum = np.log(np.abs(fourier_transform))

# Plot the magnitude spectrum
plt.imshow(magnitude_spectrum, cmap='gray')
plt.show()
```

## spatial domain filtering

Spatial domain filtering, also known as pixel-domain filtering, is called as such because it operates directly on the pixels of the image in the spatial domain, rather than transforming the image to a different domain such as the frequency domain.

Spatial domain filtering involves applying a mathematical operation to the pixels of an image in order to remove or reduce the effects of noise. This operation is typically a convolution, which is a mathematical operation that involves multiplying the values of the pixels in a neighborhood around a given pixel with a fixed set of coefficients, known as the filter kernel. The filtered value of the pixel is then computed as the sum of these products.

The filter kernel is usually designed to have certain properties that are useful for removing noise, such as preserving edges or smoothing out variations in intensity. The filter kernel is typically a small square or rectangular region of pixels, and the size of this region is known as the neighborhood size.

The spatial domain filtering approach is also called as pixel-domain filtering, because it operates directly on the pixels of the image in the spatial domain. Spatial domain filtering is simple to implement, computationally efficient, and easy to understand. It is commonly used for noise reduction in images and is widely used in various applications such as image processing, computer vision, and medical imaging

Code

```python
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt

# Read an image
gray_image = plt.imread("image_2.jpg")

# Apply a median filter to the image
filtered_image_3 = ndimage.median_filter(gray_image, size=3)
filtered_image_5 = ndimage.median_filter(gray_image, size=5)
filtered_image_7 = ndimage.median_filter(gray_image, size=7)
# Plot the original image and the filtered image
plt.figure()
plt.subplot(1, 1, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')
plt.show()
plt.subplot(1, 1, 1)
plt.imshow(filtered_image_3, cmap='gray')
plt.title('Filtered Image size 3')
plt.show()
#change size of the square neighborhood
plt.subplot(1,1, 1)
plt.imshow(filtered_image_5, cmap='gray')
plt.title('Filtered Image size 5')
plt.show()
plt.subplot(1, 1, 1)
plt.imshow(filtered_image_7, cmap='gray')
plt.title('Filtered Image size 7')
plt.show()
```

### Machine Learning

There are several machine learning based techniques that can be used to remove periodic noise from images. One such technique is to use a convolutional neural network (CNN) to learn to remove the noise from the images.

The basic idea is to train a CNN to learn to remove the noise from the images. The CNN is trained on a dataset of noisy images and corresponding clean images, where the clean images are obtained by removing the noise from the noisy images. During training, the CNN learns to recognize the patterns in the noise and learn how to remove them from the images.

Once the CNN is trained, it can be applied to new noisy images to remove the noise. The CNN can be trained using different architectures such as U-Net, ResNet, or DenseNet, depending on the specific requirements of the application.

Another technique is to use a generative model such as a Generative Adversarial Network (GAN) to learn to remove the noise from the images. A GAN consists of two neural networks: a generator and a discriminator. The generator is trained to generate clean images that are similar to the clean images in the training dataset, while the discriminator is trained to distinguish between the generated images and the clean images. During training, the generator learns to remove the noise from the images while the discriminator learns to identify the noise in the images.

In summary, Machine learning based techniques such as CNN and GAN can be used to remove periodic noise from images. These techniques can be trained on a dataset of noisy images and corresponding clean images, where the clean images are obtained by removing the noise from the noisy images. Once the model is trained, it can be applied to new noisy images to remove the noise and improve the quality of images.
